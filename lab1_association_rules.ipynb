## import packages

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## load dataset

```python
df = pd.read_csv('/content/drive/MyDrive/dsci-519-py-data/data/AmazonBinaryAll.csv')
```

## generate frequent itemset

```python
itemsets = apriori(df, min_support=0.4, use_colnames=True)
```

## generate association rules

```python
rules = association_rules(itemsets, metric="confidence", min_threshold=0.7)
```

## sort association rules by lift

```python
rules.sort_values(by=['lift'], ascending=False).head(6)
```

## print rules

```python
print(
    rules.sort_values(by=['lift'], ascending=False).head(6)
     .drop(columns=['antecedent support', 'consequent support', 'conviction'])
)
```
